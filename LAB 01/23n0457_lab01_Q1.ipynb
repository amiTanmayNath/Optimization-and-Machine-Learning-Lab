```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import ticker, cm
```

```python
def function1(x1,x2):
  return 100*x2**2 + 100*x1**4 - 200*x2*x1**2 + x1**2 - 2*x1 + 1

x = np.linspace(-4,4,1000)
y = np.linspace(-4,4,1000)

X,Y = np.meshgrid(x,y)

Z = function1(X,Y)

plt.contourf(X,Y,Z,locator = ticker.LogLocator())
plt.show()
```

```python
def function2(x1,x2):
  return x1**4 + x2**4 -20*x1**3 -20*x2**3 + 100*x1**2 +100*x2**2

x = np.linspace(-4,4,1000)
y = np.linspace(-4,4,1000)

X,Y = np.meshgrid(x,y)

Z = function2(X,Y)

plt.contourf(X,Y,Z,locator = ticker.LogLocator())
plt.show()
```

```python
def function3(x1,x2):
  return -np.cos(x1)*np.cos(x2)*np.exp(-((x1-np.pi)**2+(x2-np.pi)**2))

x = np.linspace(-4,4,1000)
y = np.linspace(-4,4,1000)

X,Y = np.meshgrid(x,y)

Z = function3(X,Y)

plt.contourf(X,Y,Z,locator = ticker.LogLocator())
plt.show()
```

```python
def function4(x1,x2):
  return 2*x1**2 - 1.05*x1**4 + (x1**6)/6 + x1*x2 + x2**2
x = np.linspace(-4,4,1000)
y = np.linspace(-4,4,1000)

X,Y = np.meshgrid(x,y)

Z = function4(X,Y)

plt.contourf(X,Y,Z,locator = ticker.LogLocator())
plt.show()
```
