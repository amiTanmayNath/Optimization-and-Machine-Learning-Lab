```python
import numpy as np
import matplotlib.pyplot as plt

tau = 1e-10
x0 = np.array([36, -49])

def gradf(x):
    return np.array([-2*x[0]-98, -2*x[1]+72])

def gradient_descent(x0, tau):
    iteration = 0
    xk = np.copy(x0)

    x1_vals, x2_vals = [x0[0]], [x0[1]]

    while norm(gradf(xk)) > tau:
        xk = xk + 0.001 * gradf(xk)
        iteration += 1
        x1_vals.append(xk[0])
        x2_vals.append(xk[1])

    return iteration, x1_vals, x2_vals

iteration, x1_vals, x2_vals = gradient_descent(x0, tau)

x1 = np.linspace(-60, 40, 400)
x2 = np.linspace(-50, 40, 400)
X1, X2 = np.meshgrid(x1, x2)
Z = g(X1, X2)

plt.contourf(X1, X2, Z, cmap='viridis')
plt.plot(x1_vals, x2_vals, marker='.', color='blue', label='Gradient Descent Trajectory', markersize=5)
plt.scatter(x0[0], x0[1], color='red', marker='o', label='Initial Point')
plt.title('Gradient Descent Trajectory')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
def f(x):
    return 256 * (x[1] - x[0]**2)**2 + (2 - x[0])**2


def gradient_f(x):
    df_dx1 = -2 * (2 - x[0]) - 512 * x[0] * (x[1] - x[0]**2)
    df_dx2 = 256 * (x[1] - x[0]**2)
    return np.array([df_dx1, df_dx2])

# Backtracking line search algorithm
def backtracking_line_search(xk, pk, alpha0, y, gamma):
    alpha = alpha0
    while f(xk + alpha * pk) > f(xk) + gamma * alpha * np.dot(gradient_f(xk), pk):
        alpha = y * alpha
    return alpha

# Gradient Descent with Backtracking Line Search
def gradient_descent_backtracking(x0, alpha0, y, gamma, max_iterations=1000, tol=1e-6):
    xk = x0
    iterations = 0
    while iterations < max_iterations:
        pk = -gradient_f(xk)
        alpha = backtracking_line_search(xk, pk, alpha0, y, gamma)
        xk = xk + alpha * pk
        iterations += 1
        if np.linalg.norm(gradient_f(xk)) < tol:
            break
    return xk, f(xk), iterations

# Starting point and parameters
x0 = np.array([100, 100])
alpha0_values = [1, 0.9, 0.75, 0.6, 0.5, 0.4, 0.25, 0.1, 0.01]
y = 0.5
gamma = 0.5

# Perform gradient descent with backtracking line search for each alpha0
results = []
for alpha0 in alpha0_values:
    minimizer, objective_value, iterations = gradient_descent_backtracking(x0, alpha0, y, gamma)
    results.append((alpha0, minimizer, objective_value, iterations))

for result in results:
    print(f"Alpha0: {result[0]}, Minimizer: {result[1]}, Objective Value: {result[2]}, Iterations: {result[3]}")

# Plot number of iterations against alpha0 values
alpha0_values_plot = [result[0] for result in results]
iterations_values_plot = [result[3] for result in results]

plt.plot(alpha0_values_plot, iterations_values_plot, marker='o')
plt.xlabel('Alpha0')
plt.ylabel('Number of Iterations')
plt.title('Number of Iterations vs Alpha0')
plt.show()
```

# Conclusions from Line Search Approaches

## 1. Performance Comparison:

The backtracking line search method usually needs fewer steps compared to the exact line search method. This is because it adjusts the step size during optimization, potentially leading to quicker convergence.

## 2. Sensitivity to Initial Step Size (\( \alpha_0 \)):

The behavior of backtracking line search depends on the initial step size (\( \alpha_0 \)). Smaller initial steps are safer but might need more steps, while larger steps can speed up the process but risk overshooting.

## 3. Convergence to the Minimizer:

Both methods reach the same solution, showing that the choice of line search doesn't affect the final result.

## 4. Effectiveness of Adaptive Step Size:

Backtracking line search adjusts the step size as needed, ensuring a balance between speed and accuracy. This is crucial for functions with varying curvature.
